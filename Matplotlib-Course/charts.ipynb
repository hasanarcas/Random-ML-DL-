# Stack Plot

```python
days = [1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.plot([], [], color="magenta", label = "Sleeping")
plt.plot([], [], color="red", label = "Eating")
plt.plot([], [], color="green", label = "Working")
plt.plot([], [], color="cyan", label = "Playing")


plt.stackplot(days, sleeping, eating, working, playing, colors=["magenta", "red", "green", "cyan"])

plt.xlabel("x")
plt.ylabel("y")
plt.title("Interesting Graph")
plt.legend()
plt.show()
```

# Subplots

```python
def create_plot():
    xs = []
    ys = []
    
    for i in range(10):
        x = i
        y = random.randrange(10)
        
        xs.append(x)
        ys.append(y)
        
    return xs,ys

ax2 = plt.subplot2grid((6,1), (0,0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((6,1), (1,0), rowspan=4, colspan=1)
ax3 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1)

x,y = create_plot()
ax2.plot(x,y, color='g')

x2,y2 = create_plot()
ax2.plot(x2,y2, color='r')

x3,y3 = create_plot()
ax3.plot(x3,y3, color='k')

plt.show()
```
