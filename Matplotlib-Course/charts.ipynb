# Legends titles and labels

```python
x = [1, 2, 3]
y = [5, 7, 4]

x2 = [1,2, 3]
y2 = [10, 14, 12]

plt.plot(x, y, label="First Line")
plt.plot(x2, y2, label="Second Line")
plt.xlabel("Plot Number")
plt.ylabel("Important Var")
plt.title("Interesting Graph")
plt.legend()
plt.show()
```

# Stack Plot

```python
days = [1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.plot([], [], color="magenta", label = "Sleeping")
plt.plot([], [], color="red", label = "Eating")
plt.plot([], [], color="green", label = "Working")
plt.plot([], [], color="cyan", label = "Playing")


plt.stackplot(days, sleeping, eating, working, playing, colors=["magenta", "red", "green", "cyan"])

plt.xlabel("x")
plt.ylabel("y")
plt.title("Interesting Graph")
plt.legend()
plt.show()
```

# Pie Chart

```python
days = [1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

slices = [7, 2, 2, 13]
activities = ["sleeping", "eating", "working", "playing"]
cols = ["m","r","g","c"]
explode= [0, 0, 0.1, 0]

plt.pie(slices, labels=activities, colors=cols,
        explode=explode, startangle=90, shadow=True,
        autopct="%1.1f")
plt.title("Interesting Graph")
plt.axis("auto")
plt.show()
```

# More Costumization

```python
df = pd.read_csv('TSLA.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df[(df.Date > pd.to_datetime('2017')) & (df.Date < pd.to_datetime('2019'))]

ax1 = plt.subplot2grid((1,1), (0,0))
ax1.plot_date(df.Date, df.Close,'-', label='Price')
price = 300
ax1.fill_between(df.Date, df.Close, price,where=(df.Close > price), facecolor='g', alpha=0.3)
ax1.fill_between(df.Date, df.Close, price,where=(df.Close<price), facecolor='r', alpha=0.3)

for label in ax1.xaxis.get_ticklabels():
    label.set_rotation(45)
    
ax1.grid(True)
ax1.xaxis.label.set_color('c')
ax1.yaxis.label.set_color('r')
# ax1.set_yticks([0,200,400,600,800])

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('TESLA STOCK PRICE')
plt.legend()
plt.subplots_adjust(left=0.09, bottom=0.18, right=0.94, top=0.9, wspace=0.2, hspace=0)
plt.show()
```

# Spines and Horizontal Lines

```python
df = pd.read_csv('TSLA.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df[(df.Date > pd.to_datetime('2017')) & (df.Date < pd.to_datetime('2019'))]

ax1 = plt.subplot2grid((1,1), (0,0))
ax1.plot_date(df.Date, df.Close,'-', label='Price')
ax1.plot([],[], linewidth=5, label='Loss', color='r', alpha=0.5)
ax1.plot([],[], linewidth=5, label='Gain', color='g', alpha=0.5)
price = 300
ax1.axhline(price, color='k', linewidth=3)
ax1.fill_between(df.Date, df.Close, price,where=(df.Close > price), facecolor='g', alpha=0.3)
ax1.fill_between(df.Date, df.Close, price,where=(df.Close<price), facecolor='r', alpha=0.3)
ax1.grid(True)

for label in ax1.xaxis.get_ticklabels():
    label.set_rotation(45)
    
ax1.set_yticks([200,250,300,350,400])
ax1.tick_params(axis='x', colors='#f06215')

ax1.xaxis.label.set_color('c')
ax1.yaxis.label.set_color('r')

ax1.spines['left'].set_color('c')
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.spines['left'].set_linewidth(5)



plt.xlabel('Date')
plt.ylabel('Price')
plt.title('TESLA STOCK PRICE')
plt.legend()
plt.subplots_adjust(left=0.09, bottom=0.18, right=0.94, top=0.9, wspace=0.2, hspace=0)
plt.show()
```

# Candlestick OHLC (open, high, low, close) Graphs

```python
df = pd.read_csv('TSLA.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df[(df.Date > pd.to_datetime('2019-01-01')) & (df.Date < pd.to_datetime('2019-03-01'))]
df.Date = mdates.date2num(df.Date)

ax1 = plt.subplot2grid((1,1), (0,0))
candlestick_ohlc(ax1, df.values, width=0.4, colorup='g', colordown='r')

for label in ax1.xaxis.get_ticklabels():
    label.set_rotation(45)
    
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax1.xaxis.set_major_locator(mticker.MaxNLocator(6))

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('TESLA STOCK PRICE')
plt.legend()
plt.subplots_adjust(left=0.09, bottom=0.18, right=0.94, top=0.9, wspace=0.2, hspace=0)
plt.show()
```
