# Adding more indicators

```python
df = pd.read_csv('TSLA.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df[(df.Date > pd.to_datetime('2018-01-01')) & (df.Date < pd.to_datetime('2019-01-01'))]
df.Date = mdates.date2num(df.Date)

style.use('fivethirtyeight')

MA1 = 10
MA2 = 30

def moving_average(values, window):
    weights = np.repeat(1.0, window)/window
    smas = np.convolve(values, weights, 'valid')
    return smas

def high_minus_low(highs, lows):
    return highs-lows

#FIRST PLOT
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=1, colspan=1)
plt.title('TESLA STOCK PRICE')
#SECOND PLOT
ax2 = plt.subplot2grid((6,1), (1,0), rowspan=4, colspan=1)
plt.xlabel('Date')
plt.ylabel('Price')
#THIRD PLOT
ax3 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1)

ma1 = moving_average(df.Close, MA1)
ma2 = moving_average(df.Close, MA2)
start = len(df.Date.iloc[MA2-1:])
h_l = list(map(high_minus_low, df.High, df.Low))

ax1.plot(df.Date, h_l, "-")

candlestick_ohlc(ax2, df.values, width=0.4, colorup='g', colordown='r')
for label in ax2.xaxis.get_ticklabels():
    label.set_rotation(45)  
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax2.xaxis.set_major_locator(mticker.MaxNLocator(6))
ax2.grid(True)

ax3.plot(df.Date.iloc[-start:], ma1[-start:])
ax3.plot(df.Date.iloc[-start:], ma2[-start:])

bbox_props = dict(boxstyle='round', fc='w', ec='k')

ax2.annotate(str(df.Close.iloc[-1]), (df.Date.iloc[-1], df.Close.iloc[-1]),
             xytext=(df.Date.iloc[-1]+6, df.Close.iloc[-1]),
             bbox=bbox_props)
 

# plt.legend()
plt.subplots_adjust(left=0.09, bottom=0.18, right=0.94, top=0.9, wspace=0.2, hspace=0)
plt.show()
```
