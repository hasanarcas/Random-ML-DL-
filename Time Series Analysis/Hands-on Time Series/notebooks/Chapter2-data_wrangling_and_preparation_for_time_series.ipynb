```python
import pandas as pd
from pandasql import sqldf
```

## Pandasql

```python
dfpsql = pd.read_excel(r'../data/Absenteeism_at_work.xls')
Query_string = ''' select * from dfpsql limit 5'''
sqldf(Query_string, globals())
```

### Applying a Filter

```python
Query_string = 'select * from dfpsql where age>=30 and age<=45'
sqldf(Query_string, globals())
```

### Distinct (Unique)

```python
Query_string = 'select distinct ID from dfpsql limit 5'
sqldf(Query_string, globals())
```

### IN method

```python
Query_string = 'select * from dfpsql where age in (20,30,40)'
sqldf(Query_string, globals())
```

### Aggregation

```python
Query_string = '''select count(Transportation_expense) as count,
                  min(Transportation_expense) as min,
                  max(Transportation_expense) as max,
                  avg(Transportation_expense) as mean
                  from dfpsql'''
sqldf(Query_string, globals())
```

### Group By

```python
Query_string = '''select ID, sum(Service_time) as Sum_Service_time
                  from dfpsql
                  group by ID
                  limit 5'''
sqldf(Query_string, globals())
```

```python
Query_string = '''select Reason_for_absence , avg(Age) as mean, min(Age) as min, max(Age) as max
                  from dfpsql
                  group by Reason_for_absence
                  limit 5'''
sqldf(Query_string, globals())
```

## Resampling

```python
df = pd.read_csv(r'../data/daily-total-female-births-CA.csv', index_col=0, parse_dates=['date'])
df.head(5)
```

### Resample by Month

```python
df.births.resample('M').mean()
```

### Resample by Quarter

```python
df.births.resample('Q').mean()
```

### Resample by Year

```python
df.births.resample('Y').mean()
```

### Resample by Week

```python
df.births.resample('W').mean()[:5]
```
